```python
from numpy import *
from sympy import *
x = symbols('x')
y = symbols('y')
```

```python
# (1) menghitung area diantara 2 kurva
f1 = x**4
f2 = 2*x - x**2
A = integrate(f2-f1, (x, 0, 1))
print(A)
```

```python
# (2) jika kurva ditukar saat pengurangan, maka hasilnya akan salah karena area tidak mungkin negatif
f1 = x**4
f2 = 2*x - x**2
A = integrate(f1-f2, (x, 0, 1))
print(A)
```

```python
# (3) menghitung area di antara 2 kurva yang lebih rumit
f1 = sqrt(4*x)
f2 = (4-4*x) / -3
A = integrate(f1-f2, (x, 0, 1))
print(A)
```

```python
# (4) integral terhadap dy
f1 = y**2
A = integrate(f1, (y, 0, 1))
print(A)
```

```python
# (5) menghitung area di antara 2 kurva yang lebih rumit, integral terhadap dy
f1 = y**2 / 4
f2 = (3*y + 4) / 4
A = integrate(f2-f1, (y, -1, 4))
print(A)
```

```python
# (6) mencari volume bangun
f1 = sqrt(x)
V = integrate(pi*(f1**2), (x, 0, 4))
print(V)
```

```python
# (7) mencari volume bangun, integral terhadap dy
f1 = cbrt(y)
V = integrate(pi*(f1**2), (y, 0, 3))
print(V)
```
